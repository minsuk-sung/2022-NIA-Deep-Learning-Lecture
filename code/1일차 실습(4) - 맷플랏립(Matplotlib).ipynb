## NumPy와 Pandas 함께 사용하기
<hr style="height:5px;border:none;color:#132e4f;background-color:#132e4f;"/>

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'x': np.arange(1000),
    'y': np.random.randint(0,100,size=1000),
    'z': np.random.randint(0,100,size=1000),
})
df
```

```python
plt.scatter(df['x'], df['y'])
```

```python
plt.scatter(df['x'], df['y'], alpha=0.5)
plt.grid(True)
```

```python
df['y'].sort_index().value_counts()[:10].plot(kind='bar')
plt.show()
```

```python
df['y'].plot(figsize=(12,8))
plt.axhline(df['y'].mean(), color='red')
plt.annotate('Mean', (500, df['y'].mean() + 1), color='red', size=20)
plt.grid(True)
```

```python
plt.matshow(df.corr())
```

```python
import seaborn as sns
corr = df.corr()
sns.heatmap(corr)
```
